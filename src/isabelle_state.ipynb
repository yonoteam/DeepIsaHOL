# Retrieving Isabelle's state

```python
from py4j.java_gateway import JavaGateway
```

```python
gateway = JavaGateway()                   # connect to the JVM
```

```python
random = gateway.jvm.java.util.Random()   # create a java.util.Random instance
number1 = random.nextInt(10)              # call the Random.nextInt method
number2 = random.nextInt(10)
print(number1, number2)
```

```python
Isabelle_System = gateway.jvm.isabelle.Isabelle_System # access the Isabelle_System scala object
Options = gateway.jvm.isabelle.Options                 # access the Isabelle Options scala object
```

```python
Isabelle_System.init("","")
```

```python
gateway.help(Options)
```

```python
gateway.help(Isabelle_System)
```

```python
gateway.shutdown()
```
